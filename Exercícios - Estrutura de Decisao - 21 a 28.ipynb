```python
'''Exercício 21 - Faça um Programa para um caixa eletrônico. 
O programa deverá perguntar ao usuário a valor do saque e depois informar quantas notas de cada valor serão fornecidas.
As notas disponíveis serão as de 1, 5, 10, 50 e 100 reais. O valor mínimo é de 10 reais e o máximo de 600 reais. 
O programa não deve se preocupar com a quantidade de notas existentes na máquina.
Exemplo 1: Para sacar a quantia de 256 reais, 
o programa fornece duas notas de 100, uma nota de 50, uma nota de 5 e uma nota de 1;
Exemplo 2: Para sacar a quantia de 399 reais, 
o programa fornece três notas de 100, uma nota de 50, quatro notas de 10, uma nota de 5 e quatro notas de 1.'''

numero = float(input('Insira o valor do saque: '))

cem = int(numero/100)
numero = numero - (cem*100)
#numero virou 56

cinquenta = int(numero/50)
numero = numero - (cinquenta*50)
#numero virou 6

dez = int(numero/10)
numero = numero - (dez*10)

cinco = int(numero/5)
numero = numero - (cinco*5)

um = numero
    
    
print('Notas 100 -',cem,)
print('Notas 50 -',cinquenta)
print('Notas 10 -',dez)
print('Notas 5 -',cinco)
print('Notas 1 -',um)
```

```python
'''Exercício 22 - Faça um Programa que peça um número inteiro e determine se ele é par ou impar. 
Dica: utilize o operador módulo (resto da divisão).
'''
numero_inteiro = int(input('Insira um numeri inteiro: '))

if numero_inteiro %2 == 0:
    print('Par')
else:
    print('Ímpar')
```

```python
'''Exercício 23 - Faça um Programa que peça um número e informe se o número é inteiro ou decimal. 
Dica: utilize uma função de arredondamento.'''

num = float(input('Numero original: '))

if num == round(num):
    print("Inteiro")
else:
    print("Decimal")
    print("Arredondado pra baixo: ", round(num-0.5) )
    print("Arredondado pra cima : ", round(num+0.5) )
        
```

```python
'''Exercício 24 -Faça um Programa que leia 2 números e em seguida pergunte ao usuário qual operação ele deseja realizar. 
O resultado da operação deve ser acompanhado de uma frase que diga se o número é:
par ou ímpar;
positivo ou negativo;
inteiro ou decimal.'''

num1 = float(input('Insira um numero: '))
num2 = float(input ('Insira outro numero: '))
operacao = input ('Qual operação deseja usar (Use: +, -, *, /): ')

if operacao == '+':
    soma = num1+num2
    print('O resultado da operação é: ',soma)
    if soma %2 == 0:
        print('Par')  
    else:
        print('Ímpar')
    if soma >= 0:
        print('Positivo')
    else:
        print('Negativo')
    if soma == round(soma):
        print('Inteiro')
    else:
        print('Decimal')

elif operacao == '-':
    subtracao = num1-num2
    print('O resultado da operação é: ',subtracao)
    if subtracao %2 == 0:
        print('Par')  
    else:
        print('Ímpar')
    if subtracao >= 0:
        print('Positivo')
    else:
        print('Negativo')
    if subtracao == round(subtracao):
        print('Inteiro')
    else:
        print('Decimal')
        
elif operacao == '*':
    multiplicacao = num1*num2
    print('O resultado da operação é: ',multiplicacao)
    if multiplicacao %2 == 0:
        print('Par')  
    else:
        print('Ímpar')
    if multiplicacao >= 0:
        print('Positivo')
    else:
        print('Negativo')
    if multiplicacao == round(multiplicacao):
        print('Inteiro')
    else:
        print('Decimal')
        
elif operacao == '/':
    divisao = num1/num2
    print('O resultado da operação é: ',divisao)
    if divisao %2 == 0:
        print('Par')  
    else:
        print('Ímpar')
    if divisao >= 0:
        print('Positivo')
    else:
        print('Negativo')
    if divisao == round(divisao):
        print('Inteiro')
    else:
        print('Decimal')

```

```python
''' Exercício 26 - Um posto está vendendo combustíveis com a seguinte tabela de descontos:
Álcool:
até 20 litros, desconto de 3% por litro
acima de 20 litros, desconto de 5% por litro
Gasolina:
até 20 litros, desconto de 4% por litro
acima de 20 litros, desconto de 6% por litro 
Escreva um algoritmo que leia o número de litros vendidos, o tipo de combustível
(codificado da seguinte forma: A-álcool, G-gasolina), calcule e imprima o valor a ser pago pelo 
cliente sabendo-se que o preço do litro da gasolina é R$ 2,50 o preço do litro do álcool é R$ 1,90.'''

litros = float(input('Insira quantidade de litros: '))
comb = str(input('Informe o tipo de combustível: A-álcool, G-gasolina - ').upper())

if comb == 'A':
    valor_sem_desconto = 1.90*litros 
    if litros <=20:
        desconto = ((3*1.90)/100)*litros
        valor_total = valor_sem_desconto-desconto
        print('Valor a ser pago:', valor_total)
    elif litros >20:
        desconto = ((5*1.90)/100)*litros
        valor_total = valor_sem_desconto-desconto
        print('Valor a ser pago:', valor_total)
        
if comb == 'G':
    valor_sem_desconto = 2.50*litros 
    if litros <=20:
        desconto = ((4*2.50)/100)*litros
        valor_total = valor_sem_desconto-desconto
        print('Valor a ser pago:', valor_total)
    elif litros >20:
        desconto = ((6*2.50)/100)*litros
        valor_total = valor_sem_desconto-desconto
        print('Valor a ser pago:', valor_total)
```
